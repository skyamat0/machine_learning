## チャレンジ問題
hint

求めたいもの:
回帰係数
$$
\hat{a} = \frac{S_{xy}}{S_x^2}, \,\, \hat{b} = \bar{y} - \hat{a} \bar{x}
$$

共分散(Covariance:Cov)，分散(Variance: Var)の求め方

$$
Cov = \frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y}) \\
Var = \frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^2
$$

配列の平均値の求め方:

`.mean()`を用いる

```python
x = np.array([1, 2, 3])
print(x.mean())
-> 2.
```

配列の分散の求め方:

`.var(ddof)`を用いる

ddof = 1で不偏分散（データ数−１で割る分散）

ddof = 0（デフォルト）で標本分散（データ数で割る分散）

```python
x = np.array([1, 2, 3])
print(x.var(ddof=0))
print(x.var(ddof=1))
->0.6666666666666666
->1.0
```

配列と配列の共分散の求め方:

`np.cov(x, y)`を用いる

返り値：
$$
\begin{pmatrix}
    S_x^2 & S_{xy}\\
    S_{yx} & S_y^2
\end{pmatrix}
$$

ddof = 1（デフォルト）で不偏分散（データ数−１で割る分散）

ddof = 0で標本分散（データ数で割る分散）
```python
x = np.array([1, 2, 3])
y = np.array([3, 7, 5])
print(np.cov(x, y, ddof=1))
print(np.cov(x, y, ddof=0))
->
[[1. 1.]
 [1. 4.]]
 [[0.66666667 0.66666667]
 [0.66666667 2.66666667]]
```
