```python
#load_bostonクラスのからbostonインスタンス生成、データの読み込み
boston = load_boston()
boston_df = pd.DataFrame(data= boston.data, columns= boston.feature_names)
boston_df["MEDV"] = boston.target
#目的変数に住宅価格
X = np.array(boston_df.drop("MEDV", axis= 1))
Y = np.array(boston_df["MEDV"])
Y = Y.reshape(-1,1)
#トレーニングデータとテストデータを分ける(作成したモデルを評価するため)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.7, test_size = 0.3, random_state = 0)
#scikitt-learnのLinearRegressionクラスからmodelインスタンス生成
model = LinearRegression()

#学習(フィッティング=パラメータを見つける)
model.fit(X_train, Y_train)
#精度を算出
print(f'訓練データ精度={model.score(X_train, Y_train)}')
print(f'テストデータ精度={model.score(X_test, Y_test)}')
print(f'偏回帰係数={model.coef_}')
print(f'定数項={model.intercept_}')
print(f'目的変数の平均値={np.average(Y_train)}')
result_df = pd.DataFrame(data= model.coef_, columns= boston.feature_names)
result_df.T
```

```python
#標準化
from sklearn import preprocessing

sscaler = preprocessing.StandardScaler()

sscaler.fit(X)

X_scaled = np.array(boston_df.drop("MEDV", axis= 1))
Y = np.array(boston_df["MEDV"])
Y = Y.reshape(-1,1)

scaled_X = sscaler.transform(X_scaled)
X_train, X_test, Y_train, Y_test = train_test_split(scaled_X, Y, train_size = 0.7, test_size = 0.3, random_state = 0)

model_scaled = LinearRegression()
model_scaled.fit(X_train, Y_train)


print(f'訓練データ精度={model_scaled.score(X_train, Y_train)}')
print(f'テストデータ精度={model_scaled.score(X_test, Y_test)}')
print(f'偏回帰係数={model_scaled.coef_}')
print(f'定数項={model_scaled.intercept_}')
print(f'目的変数の平均値{np.average(Y_train)}')
result_df = pd.DataFrame(data= model_scaled.coef_, columns= boston.feature_names)
result_df.T
```

```python
#multicollinearity
W_k = 1 - (np.sqrt(boston_df['B'])/1000 + 0.63)
W = 1000*(W_k - 0.63)
#W = np.random.rand(506, 1)
boston_df["W"] = W
X_multi = np.array(boston_df.drop("MEDV", axis= 1))
Y = np.array(boston_df["MEDV"])
Y = Y.reshape(-1,1)

sscaler_multi = preprocessing.StandardScaler()
sscaler_multi.fit(X_multi)

scaled_X_multi = sscaler_multi.transform(X_multi)
X_train, X_test, Y_train, Y_test = train_test_split(scaled_X_multi, Y, train_size = 0.7, test_size = 0.3, random_state = 0)

model_multi = LinearRegression()

model_multi.fit(X_train, Y_train)

print(f'訓練データ精度={model_multi.score(X_train, Y_train)}')
print(f'テストデータ精度={model_multi.score(X_test, Y_test)}')
print(f'偏回帰係数={model_multi.coef_}')
print(f'定数項={model_multi.intercept_}')
print(f'目的変数の平均値{np.average(Y_train)}')
result_df = pd.DataFrame(data= model_multi.coef_, columns= boston_df.drop("MEDV", axis= 1).columns)
result_df.T
```

```python
#predict
predict_X = np.array([scaled_X[0, :]])
predict_X_multi = np.array([scaled_X_multi[0, :]])
print(model_scaled.predict(predict_X))
print(model_multi.predict(predict_X_multi))
```

```python
def error(linear_model , true_y, X) -> np.ndarray:
    
    if linear_model.__class__.__name__ == "Lasso":
        pred = linear_model.predict(X).reshape(-1, 1)
        error =true_y - pred
    else:
        error = true_y - linear_model.predict(X)
    
    return error

error_base = error(model, Y, X)
error_scaled = error(model_scaled, Y, scaled_X)
error_multi = error(model_multi, Y, scaled_X_multi)

sigma_2 = error_base.T @ error_base
sigma_2_scaled = error_scaled.T @ error_scaled
sigma_2_multi = error_multi.T @ error_multi

cov_matrix = sigma_2 * np.linalg.inv(X.T @ X)
scaled_cov_matrix = sigma_2_scaled * np.linalg.inv(scaled_X.T @ scaled_X)
multi_cov_matrix = sigma_2_multi * np.linalg.inv(scaled_X_multi.T @ scaled_X_multi)
print(np.trace(cov_matrix))
print(np.trace(scaled_cov_matrix))
print(np.trace(multi_cov_matrix))
```
