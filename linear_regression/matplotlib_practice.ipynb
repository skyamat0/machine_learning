```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#X:インスタグラムの投稿数 y:年間のケーキの売り上げ(万円)
X = np.array([5, 3, 6, 11, 7, 15, 20, 1])
y = np.array([240,  110, 240, 600, 210, 580, 700, 80])
plt.scatter(X, y, color='y', marker='x', s=500,)
plt.xlim(0,30)
plt.ylim(0, 1000)
plt.xlabel("num of posts in instagram")
plt.ylabel("earnings")
plt.title("graph 1")
plt.show()
```

```python
#傾きa
a = 3
#切片b
b = 1
x = np.arange(-2, 2)
y = a*x+b
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
X = np.array([5, 3, 6, 11, 7, 15, 20, 1])
y = np.array([240,  110, 240, 600, 210, 580, 700, 80])
df = pd.DataFrame(data=[X, y]).T
df = df.rename(columns={0: "インスタ投稿数", 1:"売り上げ"})
df
```

```python
#DataFrameからもイケる
X = df["インスタ投稿数"]
y = df["売り上げ"]
plt.scatter(X, y)
plt.xlim(0,30)
plt.ylim(0, 1000)
plt.xlabel("num of posts in instagram")
plt.ylabel("earnings")
plt.show()
```
