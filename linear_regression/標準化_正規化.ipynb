# 標準化とは
データ$x$から平均値を引いて，標準偏差で割るという操作のこと．

これによって，データの平均を０に，標準偏差を１にすることができる．

標準化
$$
x_{z} = \frac{x-\bar{x}}{\sigma_x}
$$

```python
import matplotlib.pyplot as plt
import numpy as np
def normal(x, mu, sigma):
    return np.exp(-(x-mu)**2/(np.sqrt(2*np.pi*(sigma**2))))

_x = np.random.normal(loc=15, scale=np.sqrt(2), size=100)
_x_st = (_x - _x.mean())/_x.std()

plt.hist(_x_st)
plt.hist(_x)
```

# 正規化とは
データ$x$から最小値を引いて，（最大値ー最小値）で割るという操作のこと．

これによって，データの最小値を０に，最大値を１にすることができる．

正規化
$$
x_{n} = \frac{x-x_{min}}{x_{max}-x_{min}}
$$

```python
_x_nor = (_x - _x.min())/(_x.max() - _x.min())
plt.hist(_x_nor)
plt.hist(_x)
```

## どちらが良い？？

基本的には標準化．

どちらも単位の影響度を無視できるようにする変換であるが，

正規化は，最大値と最小値が定まっている必要がある，外れ値に弱いというデメリットがあるためである．
