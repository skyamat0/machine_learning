```python
from sklearn.datasets import make_moons
moons = make_moons(n_samples=100,
                   random_state=0, 
                   noise=0.3
                   ) 
```

```python
import matplotlib.pyplot as plt
```

```python
#moons_plot
X = moons[0]
y = moons[1]
plt.scatter(X[y == 0, 0], X[y == 0, 1],c='r', marker='o')
plt.scatter(X[y == 1, 0], X[y == 1, 1],c='b', marker='^')
plt.show()
```
