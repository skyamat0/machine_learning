```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

cols_name = np.array(["定期テストの平均点", "出席率(%)"])
target_name = "志望校に合格できたか"
X = np.array([[82, 100],
              [38, 78],
              [50, 90],
              [62, 93],
              [47, 93],
              [52, 85],
              [93, 96],
              [58, 86],
              [67, 75],
              [80, 92]])
y = np.array([1, 0, 0, 1, 0, 1, 1, 1, 1, 1])
```

```python
#dataframeに格納してみよう
#hint: (columns=cols_name, data=X)
#hint: df[target_name] = y
df = pd.DataFrame(X, columns=cols_name)
df[target_name] = y
```

```python
df
```

```python
#散布図を書いてみよう.ただし,y=0の時はblueにy=1の時はredになるように描画すること
plt.scatter(X[y==0, 0], X[y==0, 1], c="b")
plt.scatter(X[y==1, 0], X[y==1, 1], c="r")
plt.show()
```

```python
#上記散布図に加え、以下のデータを星印にして描画してみよう.
#星印のデータはサイズをs=200に設定すること
unknown = np.array([55, 97])

plt.scatter(X[y==0, 0], X[y==0, 1], c="b")
plt.scatter(X[y==1, 0], X[y==1, 1], c="r")
plt.scatter(unknown[0], unknown[1], c="g", marker="*", s=200)
plt.show()
```

```python
#星印の人は合格できるだろうか？
#k近傍法(k=1)で予測してみよう
#できたら、k=2の時,k=3の時も考えてみよう
#hint: 図的に近い点を探して、その点との距離を測って、近いものを予測値とする
d1 = np.linalg.norm(unknown-np.array([50, 90]))
d2 = np.linalg.norm(unknown-np.array([62, 93]))
d3 = np.linalg.norm(unknown-np.array([47, 93]))
print(d1)
print(d2)
print(d3)
```
