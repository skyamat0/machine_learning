```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_circles
```

```python
circles = make_circles(noise=0.05)
X, y = circles[0], circles[1]
plt.scatter(X[y==0, 0], X[y==0, 1], c="r")
plt.scatter(X[y==1, 0], X[y==1, 1], c="b")
plt.show()
```

```python
#描画領域の確保(一番外側の点＋aの点を取る)
x_min, x_max = X[:, 0].min() - 0.1, X[:,0].max() + 0.1
y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1

plt.scatter(X[y==0, 0], X[y==0, 1], c="r")
plt.scatter(X[y==1, 0], X[y==1, 1], c="b")
plt.scatter(x_min, y_min, s=200, c="g")
plt.scatter(x_min, y_max, s=200, c="g")
plt.scatter(x_max, y_min, s=200, c="g")
plt.scatter(x_max, y_max, s=200, c="g")
plt.show()
```

```python
#np.meshgrid()は引数の配列から格子点を作り出すメソッド
#np.linspace()は引数の値から[第三引数]個の等間隔の値を作り出すメソッド
#以下では、x_min~x_max, y_min~y_maxのそれぞれの範囲で5個ずつ点を作り、それぞれで格子点を作り出している
print(np.linspace(1, 2, 5))
print(np.linspace(1, 2, 5))
#-> (upper_bound - lower_bound) / (n-1)の間隔での数値列が与えられる
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 5), np.linspace(y_min, y_max, 5))
plt.scatter(xx, yy)
plt.show()
```

```python
#さらに細かく分割する
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 10), np.linspace(y_min, y_max, 10))
plt.scatter(xx, yy)
plt.show()
```

```python
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 50), np.linspace(y_min, y_max, 50))
plt.scatter(xx, yy)
plt.show()
```

```python
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
plt.scatter(xx, yy)
plt.show()
```

```python
#モデル作成
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X, y)
```

```python
#つくった格子点のモデルによる予測値を合わせて描画する
#格子点数 = 25
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 5), np.linspace(y_min, y_max, 5))
X_input = np.c_[xx.flatten(), yy.flatten()]
print(X_input.shape)
print(X.shape)
# X_input = np.vstack([xx.flatten(), yy.flatten()]).T
# X_input = np.hstack([xx.flatten().reshape(-1, 1), yy.flatten().reshape(-1,1)])
# いずれも同じ結果
# np.flatten() vs np.ravel()
# flattenは元の引数をcopyする非破壊処理、ravelは元の引数を書き換える破壊処理。後者の方がメモリ節約になり、速いが引数が書き変わることに注意

y_pred = model.predict(X_input)
plt.scatter(X_input[y_pred==0, 0], X_input[y_pred==0, 1], c="r")
plt.scatter(X_input[y_pred==1, 0], X_input[y_pred==1, 1], c="b")
plt.show()
```
