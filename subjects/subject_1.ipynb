```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_linnerud
```

```python
linnerud = load_linnerud()
linnerud_df = pd.DataFrame(data = linnerud.data, columns = linnerud.feature_names) 
#説明変数：懸垂の回数(chins),腹筋の回数(situps),跳躍(jumps) 
#目的変数 : 体重,ウエスト,脈拍
linnerud_df["Weight"] = linnerud.target[:,0]
linnerud_df
```

```python
X = linnerud['data'][:,0].reshape(-1,1)
Y = linnerud['target'][:,0].reshape(-1,1)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.7, test_size = 0.3, random_state = 0)
```

```python
fig, ax = plt.subplots()
plt.figure(figsize=(10,10))
ax.scatter(X, Y, c='b')
ax.set_title('Chin-ups and weight dependency preview')
ax.set_xlabel('Chin-ups')
ax.set_ylabel('Weight')
plt.show()
```

```python
model = LinearRegression()

model.fit(X_train, Y_train)

Y_pred = model.predict(X_test)
```

```python
fig, ax = plt.subplots(figsize=(5, 5), dpi=100)
ax.scatter(X_train, Y_train, c='b')
ax.scatter(X_test, Y_test, c='g')
ax.plot(X_test, Y_pred)

ax.set_title('Simulated data and estimated line for Chins vs Weight')
ax.set_xlabel('Chins')
ax.set_ylabel('Weight')
plt.show()
```

```python
print(f'訓練データ精度={model.score(X_train, Y_train)}')
print(f'テストデータ精度={model.score(X_test, Y_test)}')
print(f'偏回帰係数={model.coef_}')
print(f'定数項={model.intercept_}')
```
