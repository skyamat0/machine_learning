```python
df_2015 = df[df["西暦（年）"] == 2015]
df_2015 = df_2015.set_index("都道府県名").loc["北海道":]

df2 = df2.rename(columns={"全国，都道府県，市区町村": "都道府県", "面積（参考）【km2】":"面積"})
df2 = df2.loc[:, ["都道府県", "面積"]]
df2 = df2[df2["都道府県"].isin(todoufuken_parser.keys())]
df2 = df2.set_index("都道府県").loc["北海道":]
```

```python
change = {}
for s in np.array(df_2015["面積"]):
    tmp = s
    s = s.replace(",", "")
    change[tmp] = float(s)
```

```python
df_2015 = df_2015.replace(change)
```

##都道府県の人口と面積でクラスタリングせよ

>結果をデンドログラムで出力すること

```python
df_2015.head()
```

```python
#----------------------------------code----------------------------------#
X, y = df_2015.loc[:, ["人口（総数）", "面積"]], df_2015.index
def plot_dendrogram(model, ax, labels, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack([model.children_, model.distances_,
                                      counts]).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, ax=ax, labels=labels, **kwargs)

model_ward = AgglomerativeClustering(n_clusters=None, 
                                affinity="euclidean",
                                distance_threshold=0,
                                linkage="ward")
model_ward.fit(X)
fig, ax = plt.subplots(figsize=(10, 5))
plot_dendrogram(model_ward, ax=ax, labels=y, truncate_mode='level', p=3)
fig.show()
#----------------------------------//code--------------------------------#
```

```python
X = np.array(X.astype(float))
fig, ax = plt.subplots(figsize=(20, 10))
linkage_arr = ward(X)
dendrogram(linkage_arr, labels=y, ax=ax)
fig.show()
```

```python
model = AgglomerativeClustering(n_clusters=3)
model.fit(X)
result = model.fit_predict(X)
print(result)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = AgglomerativeClustering(n_clusters=5)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X, model.fit_predict(X))
data = {"東京":[9126214, 2416.11], "大阪":[8839469, 1905.32], "鳥取":[573441, 3507.14]}
print(knn.predict(np.array([data["東京"],
                            data["大阪"],
                            data["鳥取"]])))
```
