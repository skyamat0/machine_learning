```python
from scipy.cluster.hierarchy import dendrogram, ward, linkage
import numpy as np
import matplotlib.pyplot as plt

apartments = {'A':np.array([1, 3]),
             'B':np.array([2, 2]),
             'C':np.array([3, 4]),
             'D':np.array([7, 9]),
             'E':np.array([8, 7]),
             'F':np.array([9, 9])}
arr = np.array([apartments['A'], apartments['B'], apartments['C'], apartments['D'], apartments['E'], apartments['F']])


linkage_arr = linkage(arr, "single")
print(linkage_arr)
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
dendrogram(linkage_arr, labels=list(apartments.keys()), ax=ax[0])
ax[1].scatter(arr[:,0], arr[:,1])
fig.show()
```
