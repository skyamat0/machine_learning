```python
from scipy.cluster.hierarchy import dendrogram, ward
import numpy as np
import matplotlib.pyplot as plt
```

```python
fruits = {"apple": np.array([200, 10]),
          "orange": np.array([100, 30]),
          "banana": np.array([200, 20]),
          "melon": np.array([1000, 3]),
          "w_melon": np.array([980, 10])
         }
fruits_matrix = np.array([fruits[fruit] for fruit in fruits])
fruits_list = fruits.keys()
fruits_list = np.array(list(fruits_list))
print(np.array(list(fruits_list)))
print(fruits_matrix)
```

```python
linkage_arr = ward(fruits_matrix)
print(linkage_arr)
```

```python
fig, ax = plt.subplots()
dendrogram(linkage_arr, labels=fruits_list, ax=ax)
fig.show()
```

```python
#apartmentsでデンドログラムを作成せよ
apartments = {'A':np.array([1, 3]),
             'B':np.array([2, 2]),
             'C':np.array([3, 4]),
             'D':np.array([7, 9]),
             'E':np.array([8, 7]),
             'F':np.array([9, 9])}
#------------------code--------------------#
arr = np.array([apartments['A'], apartments['B'], apartments['C'], apartments['D'], apartments['E'], apartments['F']])
linkage_arr = ward(arr)
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
dendrogram(linkage_arr, labels=list(apartments.keys()), ax=ax[0])
ax[1].scatter(arr[:,0], arr[:,1])
fig.show()
#------------------//code------------------#
```
