```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_boston
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(load_boston().data, load_boston().target, random_state=0)
gbr = GradientBoostingRegressor()
gbr.fit(X_train, Y_train)
```

```python
print(f"train accuracy: {gbr.score(X_train, Y_train)}")
print(f"test accuracy: {gbr.score(X_test, Y_test)}")
```

```python
optimized_gbr = GradientBoostingRegressor()
optimized_gbr.fit(X_train, Y_train)
print(f"gbr train accuracy: {optimized_gbr.score(X_train, Y_train)}")
print(f"gbr test accuracy: {optimized_gbr.score(X_test, Y_test)}")
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
cv = KFold(n_splits=3, shuffle=True, random_state=42) #k分割交差検証
cv_params = {"learning_rate": [i / 100 for i in range(1, 10, 1)],
             "n_estimators": list(range(100, 200, 10)),
             "max_depth": list(range(1, 4, 1))}
gridcv = GridSearchCV(optimized_gbr, param_grid=cv_params, cv=cv,
                      scoring="r2", n_jobs=-1)
gridcv.fit(X_train, Y_train)
best_params = gridcv.best_params_
best_score = gridcv.best_score_
print(f'best parameter {best_params}')

optimized_gbr = GradientBoostingRegressor(learning_rate=best_params["learning_rate"], n_estimators=best_params["n_estimators"], max_depth=best_params["max_depth"])
optimized_gbr.fit(X_train, Y_train)
print(f"optimized train accuracy: {optimized_gbr.score(X_train, Y_train)}")
print(f"optimized test accuracy: {optimized_gbr.score(X_test, Y_test)}")
```
