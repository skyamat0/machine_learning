```python
#subject3
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
#import mglearn
```

```python
X, Y = make_moons(n_samples=100, noise=0.3, random_state=0)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=0)
```

```python
forest = RandomForestClassifier(n_estimators=100, max_depth=None, max_features="auto", max_samples=None, random_state=0)
tree = DecisionTreeClassifier(max_depth=None, max_features=None,random_state=0)
forest.fit(X_train, Y_train)
tree.fit(X_train, Y_train)
print("random forest train score: ", format(forest.score(X_train, Y_train)))
print("random forest test score: ", format(forest.score(X_test, Y_test)))
print("=============================")
print("decision tree train score: ", format(tree.score(X_train, Y_train)))
print("decision tree test score: ", format(tree.score(X_test, Y_test)))
```

```python
fig, axes = plt.subplots(2, 3, figsize=(20, 10))
for i, (ax, tree) in enumerate(zip(axes.ravel(), forest.estimators_)):
  ax.set_title(f"Title {i}")
  mglearn.plots.plot_tree_partition(X_train, Y_train, tree, ax=ax)
mglearn.plots.plot_2d_separator(forest, X_train, fill=True, ax=axes[-1, -1], alpha=0.4)
axes[-1, -1].set_title("Random Forest")
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], Y_train)
list(enumerate(zip(axes.ravel(), forest.estimators_)))
```

```python
powerful_forest = RandomForestClassifier(n_estimators=100, random_state=0)
powerful_forest.fit(X_train, Y_train)
```

```python
fig, axes = plt.subplots(2, 3, figsize=(20, 10))
print(X_train)
for i, (ax, tree) in enumerate(zip(axes.ravel(), powerful_forest.estimators_)):
  ax.set_title("Title {}".format(i))
  mglearn.plots.plot_tree_partition(X_train, Y_train, tree, ax=ax)
mglearn.plots.plot_2d_separator(powerful_forest, X_train, fill=True, ax=axes[-1, -1], alpha=0.4)
axes[-1, -1].set_title("Random Forest")
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], Y_train)
```

```python
print(f"10 trees random forest: train accuracy = {forest.score(X_train, Y_train)}")
print(f"10 trees random forest: test accuracy = {forest.score(X_test, Y_test)}")
print(f"100 trees random forest: train accuracy = {powerful_forest.score(X_train, Y_train)}")
print(f"100 trees random forest: test accuracy = {powerful_forest.score(X_test, Y_test)}")
```

```python
predict = powerful_forest.predict(X)
print(predict)
print(Y)
error_count = 0
for i in range(len(Y)):
  if predict[i] != Y[i]:
    error_count += 1
print(error_count)
```
