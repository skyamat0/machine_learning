```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv("data.csv", header=0)
df.head()
```

```python
items = set()
for col in df:
    items.update(df[col].unique())
itemset = set(items)
encoded_vals = []
for index, row in df.iterrows():
    rowset = set(row) 
    labels = {}
    uncommons = list(itemset - rowset)
    commons = list(itemset.intersection(rowset))
    for uc in uncommons:
        labels[uc] = 0
    for com in commons:
        labels[com] = 1
    encoded_vals.append(labels)
encoded_vals[0]
ohe_df = pd.DataFrame(encoded_vals)
ohe_df = ohe_df.drop(columns=ohe_df.columns[0], axis=1)
```

```python
ohe_df
```

```python
recommended_rules = apriori(ohe_df, min_support=0.2, use_colnames=True)
recommended_rules
```

```python
association_rules(recommended_rules, metric="confidence", min_threshold=0.4)
```
