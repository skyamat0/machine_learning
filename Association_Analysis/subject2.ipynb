# setオブジェクト
集合のオブジェクト
```
x = set()
y = {"a","b","c"}
```
のように記述する

集合に関する操作ができる

ex)
```python
A = {"a", "b", "c"}
B = {"a", "b", "d"}
#AまたはB
A.unioun(B)
A | B
-> {"a", "b", "c", "d"}
#AかつB
A.intersection(B)
A & B
-> {"a", "b"}
#Aのみに入っているもの
A.difference(B)
A - B
->{(空集合)}
#AがBの部分集合
A.issubset(B)
A <= B
True
#AがBの真部分集合
A < B
True
```

など。。。

```python
#要素の追加
A.add("e")
-> {"a", "b", "c", "e"}
#要素の削除
A.remove("b")
-> {"a", "c", "e"}
#要素の取り出し
A.pop()
```
