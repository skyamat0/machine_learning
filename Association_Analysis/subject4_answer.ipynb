```python
transactions = [{"パン","加工肉製品","ジャム", "牛乳","卵"},
                {"パン","オムツ","ビール","加工肉製品","カップ麺"},
                {"弁当","ビール"},
                {"ビール","枝豆","チューハイ","弁当","オムツ"},
                {"肉","野菜","レトルトカレー","オムツ","ビール"},
                {"パン","オムツ","離乳食","牛乳","卵"}]
def support(X, transactions):
    count = 0
    for transaction in transactions:
        if X <= transaction:
            count += 1
    return count / len(transactions)

def confidence(X, Y):
    X_Y = X | Y
    return support(X_Y, transactions)/support(X, transactions)

def lift(X, Y):
    return confidence(X, Y) / support(Y, transactions)

X1 = {"パン"}
Y1 = {"加工肉製品"}
print(lift(X1, Y1))

X2 = {"オムツ"}
Y2 = {"ビール"}
print(lift(X2, Y2))

X3, Y3 = {"パン"}, {"卵"}
print(lift(X3, Y3))
```
