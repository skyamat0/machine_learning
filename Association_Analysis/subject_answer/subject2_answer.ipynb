```python
transactions = [{"パン","加工肉製品","ジャム", "牛乳","卵"},
                {"パン","オムツ","ビール","加工肉製品","カップ麺"},
                {"弁当","ビール"},
                {"ビール","枝豆","チューハイ","弁当","オムツ"},
                {"肉","野菜","レトルトカレー","オムツ","ビール"},
                {"パン","オムツ","離乳食","牛乳","卵"}]
def support(X, transactions):
    count = 0
    for transaction in transactions:
        if X <= transaction:
            count += 1
    return count / len(transactions)
print(support({"パン","加工肉製品"}, transactions))
print(support({"オムツ","ビール"}, transactions))
print(support({"パン","卵"}, transactions))
```
