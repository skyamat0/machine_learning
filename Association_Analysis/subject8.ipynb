```python
import mlxtend
```

```python
transactions = [["パン","加工肉製品","ジャム", "牛乳","卵"],
                ["パン","オムツ","ビール","加工肉製品","カップ麺"],
                ["弁当","ビール"],
                ["ビール","枝豆","チューハイ","弁当","オムツ"],
                ["肉","野菜","レトルトカレー","オムツ","ビール"],
                ["パン","オムツ","離乳食","牛乳","卵"]]
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd
```

```python
te = TransactionEncoder()
te_arr = te.fit(transactions).transform(transactions)
df = pd.DataFrame(data=te_arr, columns=te.columns_)
df
```

```python
apriori(df, min_support=0.5, use_colnames=True)
```

```python
freq_items = apriori(df, min_support=0.5, use_colnames=True)
association_rules(freq_items, metric="confidence", min_threshold=0.6)
```
