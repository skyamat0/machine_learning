```python
import numpy as np
import matplotlib.pyplot as plt
import mglearn
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
x, y = make_blobs(n_samples=200, random_state=10,centers=5, cluster_std=[1.5, 1.5, 1.5, 1.5, 1.5])
x, y = make_blobs(n_samples=200, random_state=10,centers=5, cluster_std=[1.0, 3.0, 1.0, 2.5, 3.0])
mglearn.discrete_scatter(x[:,0], x[:,1], y)
```

```python
clusters = np.arange(1, 10)
fig, ax = plt.subplots()
for i in clusters:
    kmeans = KMeans(n_clusters=i, 
                    init='k-means++',
                    random_state=10)
    kmeans.fit(x)
    ax.scatter(i, kmeans.inertia_)
ax.set_xlabel('Number of clusters')
ax.set_ylabel('Distortion')
fig.show()
```
