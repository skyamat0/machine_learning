```python
import numpy as np
import matplotlib.pyplot as plt
import mglearn
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
#データ生成
X, y = make_blobs(n_samples=200, random_state=10, centers=5, cluster_std=[1.5, 1.5, 1.5, 1.5, 1.5])
```

```python
#kmeansモデルを作成し、kmeans.inertia_属性によりクラスター内平方和を確認せよ
kmeans = KMeans(n_clusters=5, init="k-means++")
kmeans.fit(X)
print(kmeans.inertia_)
```

```python
#クラスター数(kの数)を変化させて、クラスター内平方和をそれぞれ計算し、同時に散布図プロットせよ
#簡単なやり方は、for文内部でkmeansを毎回学習させて、kmeans.inertia_をプロットしていくのが簡単
for i in range(1,10):
    kmeans = KMeans(n_clusters=i, init="k-means++")
    kmeans.fit(X)
    plt.scatter(i, kmeans.inertia_)
plt.show()
```
