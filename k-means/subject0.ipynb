```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Apple = np.array([200, 10])
Orange = np.array([100, 30])
Banana = np.array([200, 20])
Melon = np.array([1000, 3])
WaterMelon = np.array([980, 10])
```

重心計算：データを成分ごとに全部足して、その数で割る

```python
#りんごとバナナとみかんのデータの平均(重心)を計算せよ
ave1 = (Apple + Banana + Orange)/3
print(ave1)
```

```python
#りんごとバナナとみかん、メロン、スイカのデータの平均(重心)を計算せよ
ave2 = (Apple + Banana + Orange + Melon + WaterMelon)/5
print(ave2)
```

```python
#重心と各フルーツのデータを散布図として描画せよ
l = [Apple, Banana, Orange, Melon, WaterMelon]
for i in l:
    plt.scatter(i[0], i[1])
plt.scatter(ave1[0], ave1[1], marker="*", s=200)
plt.scatter(ave2[0], ave2[1], marker="*", s=200)
plt.show()
```

距離計算：2点間の距離の公式を使うか、np.dot(A,A.transpose())を用いる

または、np.linalg.norm(x-y)

```python
#重心とりんごのデータの距離を計算せよ
d = ave2 - Apple
print(d)
d = np.sqrt(d[0]**2 + d[1]**2)
print(d)
d = np.sqrt(np.dot(d,d.transpose()))
print(d)
d = np.linalg.norm(d)
print(d)
```

```python
#各フルーツと重心との距離を計算せよ
fruits = [Apple, Banana, Orange, Melon, WaterMelon]
for fruit in fruits:
    d = np.linalg.norm(ave2-fruit)
    print(d)
```
