```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def distance(vector1, vector2):
    v = vector1 - vector2
    return np.sqrt(np.dot(v, np.transpose(v)))
```

```python
apartment = {'A':np.array([1, 3]), 
             'B':np.array([2, 2]), 
             'C':np.array([3, 4]), 
             'D':np.array([7, 9]), 
             'E':np.array([8, 7]), 
             'F':np.array([9, 9])}
euclidean_distance = {}
for name1 in apartment:
    tmp = np.empty((0))
    for name2 in apartment:
        d = distance(apartment[name1], apartment[name2])
        tmp = np.append(tmp, d)
    euclidean_distance[name1] = tmp

df1 = pd.DataFrame(apartment).transpose()
df1.columns = ["築年数", "所要時間"]
df2 = pd.DataFrame(euclidean_distance, index=[name for name in apartment])
df2.to_csv("apartment_distance.csv")
print(df2)
print(df1)
```
