```python
import numpy as np
import matplotlib.pyplot as plt
import pandas
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
Apple = np.array([2000, 10])
Orange = np.array([1000, 30])
Banana = np.array([2000, 20])
Melon = np.array([10000, 3])
WaterMelon = np.array([9800, 10])
fruits = np.array([Apple, Orange, Banana, Melon, WaterMelon])
```

```python
plt.scatter(fruits[:, 0], fruits[:, 1])
plt.show()
```

```python
km = KMeans(n_clusters=2, init="k-means++")
km.fit(fruits)
print(km.cluster_centers_)
```

```python
plt.scatter(fruits[:, 0], fruits[:, 1])
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1])
plt.show()
```

```python
x_min, x_max = fruits[:, 0].min() - 0.1, fruits[:,0].max() + 0.1
y_min, y_max = fruits[:, 1].min() - 0.1, fruits[:, 1].max() + 0.1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
X_input = np.c_[xx.flatten(), yy.flatten()]
y_pred = km.predict(X_input)
plt.scatter(X_input[y_pred==0, 0], X_input[y_pred==0, 1], c="r", alpha=0.05)
plt.scatter(X_input[y_pred==1, 0], X_input[y_pred==1, 1], c="b", alpha=0.05)
plt.scatter(fruits[:, 0], fruits[:, 1])
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1])
plt.show()
```

```python
print(np.std(np.sqrt(fruits[:, 0]**2+fruits[:, 1]**2))) #距離
print(np.std(fruits[:, 0])) #金額
print(np.std(fruits[:, 1])) #売上個数
```

```python
#標準化
scaler = StandardScaler()
scaled_fruits = scaler.fit_transform(fruits)
km_scaled = KMeans(n_clusters=2, init="k-means++")
km_scaled.fit(scaled_fruits)
print(km_scaled.cluster_centers_)
```

```python
plt.scatter(fruits[:, 0], fruits[:, 1])
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1])
plt.show()
```

```python
plt.scatter(scaled_fruits[:, 0], scaled_fruits[:, 1])
plt.scatter(km_scaled.cluster_centers_[:, 0], km_scaled.cluster_centers_[:, 1])
plt.show()
```

```python
x_min, x_max = scaled_fruits[:, 0].min() - 0.1, scaled_fruits[:,0].max() + 0.1
y_min, y_max = scaled_fruits[:, 1].min() - 0.1, scaled_fruits[:, 1].max() + 0.1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
X_input = np.c_[xx.flatten(), yy.flatten()]
y_pred = km_scaled.predict(X_input)
plt.scatter(X_input[y_pred==0, 0], X_input[y_pred==0, 1], c="r", alpha=0.05)
plt.scatter(X_input[y_pred==1, 0], X_input[y_pred==1, 1], c="b", alpha=0.05)
plt.scatter(scaled_fruits[:, 0], scaled_fruits[:, 1])
plt.scatter(km_scaled.cluster_centers_[:, 0], km_scaled.cluster_centers_[:, 1])
plt.show()
```

```python
print(np.std(np.sqrt(scaled_fruits[:, 0]**2+scaled_fruits[:, 1]**2)))
print(np.std(scaled_fruits[:, 0])) #amount
print(np.std(scaled_fruits[:, 1])) #count
```
