```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import mglearn
import numpy as np
import matplotlib.pyplot as plt
```

```python
#データ生成
X, y = make_blobs(random_state=10, centers=5)
kmeans = KMeans(n_clusters=5, init='k-means++',)
kmeans.fit(X)
```

```python
#KMeansからインスタンスを生成し、fitメソッドで学習させよ
#※教師なし学習なので,Xのみfitメソッドに渡せば良いことに注意
```

```python
#kmeans.labels_属性とkmeans.cluster_centers_属性を確認し、ラベルわけされた結果とクラスタ中心を確認せよ
print(kmeans.labels_)
print(kmeans.cluster_centers_)
```

```python
#少し考えて、クラスタ中心とラベルで色分けされたデータXを散布図描画せよ
for i in np.unique(kmeans.labels_):
    plt.scatter(X[kmeans.labels_ == i, 0], X[kmeans.labels_ == i, 1])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=500, marker="*")
plt.show()
```

```python
clusters = np.arange(1, 10)
fig, ax = plt.subplots()
for i in clusters:
    kmeans = KMeans(n_clusters=i, 
                    init='k-means++',
                    random_state=10)
    kmeans.fit(X)
    ax.scatter(i, kmeans.inertia_)
ax.set_xlabel('Number of clusters')
ax.set_ylabel('Distortion')
fig.show()
```
