```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Apple = np.array([200, 10])
Orange = np.array([100, 30])
Banana = np.array([200, 20])
Melon = np.array([1000, 3])
WaterMelon = np.array([980, 10])
```

重心計算：データを成分ごとに全部足して、その数で割る

```python
#りんごとバナナとみかんのデータの平均(重心)を計算せよ
cog = (Apple + Orange + Banana)/5
print(cog)
```

```python
#りんごとバナナとみかん、メロン、スイカのデータの平均(重心)を計算せよ
cog = (Apple + Orange + Banana + Melon + WaterMelon)/5
print(cog)
```

```python
#参考.numpyを用いて、行列形式のまま計算する
fruits = np.array([Apple,
                Orange,
                Banana,
                Melon,
                WaterMelon])
print(fruits)
print(fruits.shape)
```

```python
#行を足し合わせ
cog = np.sum(fruits, axis=0)/fruits.shape[0]
cog
```

```python
#重心と各フルーツのデータを散布図として描画せよ
plt.scatter(fruits[0:, 0], fruits[0:, 1])
plt.scatter(cog[0], cog[1])
```

距離計算：2点間の距離の公式を使うか、np.dot(A,A.transpose())を用いる

または、np.linalg.norm(x-y)

```python
#重心とりんごのデータの距離を計算せよ

#(X-Y)(X-Y)tで重心との距離計算
d_Apple = np.sqrt(
    np.dot(
    (Apple - cog), 
    (Apple - cog).transpose()
))
print(d_Apple)
#d^2 = (x2-x1)^2+(y2-y1)^2で重心との距離計算
d_Apple = np.sqrt(
    np.sum(
    (Apple - cog)*(Apple - cog),
    axis=0)
)

d_Apple = np.linalg.norm(Apple - cog)
#print(Apple - cog)
#print((Apple - cog)*(Apple - cog))
print(d_Apple)
```

```python
#各フルーツと重心との距離を計算せよ
d_Orange = np.sqrt(
    np.sum(
        (Orange - cog) * (Orange - cog),
        axis=0
    )
)
print(d_Orange)
d_Banana = np.sqrt(
    np.sum(
        (Banana - cog) * (Banana - cog),
        axis=0
    )
)
print(d_Banana)
d_Melon = np.sqrt(
    np.sum(
        (Melon - cog) * (Melon - cog),
        axis=0
    )
)
print(d_Melon)
d_WaterMelon = np.sqrt(
    np.sum(
        (WaterMelon - cog) * (WaterMelon - cog),
        axis=0
    )
)
print(d_WaterMelon)
```

```python
#一気に計算(参考までに)
matrix_cog = np.array([cog,cog,cog,cog,cog])
d_fruits = np.sqrt(
    np.dot(
        (fruits - cog),
        (fruits - cog).transpose()
    )
)
print(d_fruits)
```
