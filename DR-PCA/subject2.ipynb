```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

def component_vector(arr, pca_model, ax):
    x = np.array([-5, 5])
    g = np.sum(arr, axis=0)/arr.shape[0]
    gradient = pca_model.components_[0][1]/pca_model.components_[0][0]
    intercept = g[1] - gradient*g[0]
    y = gradient*x+intercept
    ax.plot(x, y)
    return np.array([x, y])

mean_height = 173
mean_weight = 68
x = np.random.normal(loc=173.0, scale=5.0, size=(1,100))
y = np.random.normal(loc=68.0, scale=6.0, size=(1,100))
data = np.array([x[0],y[0]]).transpose()
sscaler = StandardScaler()
sscaler.fit(data)
data_scaled = sscaler.transform(data)
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.set_xlim(150, 200)
ax.set_ylim(40, 100)
```
