```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from mpl_toolkits.mplot3d import Axes3D
```

```python
n=100
X, y = make_circles(n_samples=n, factor=0.2, noise=0.15, )
```

```python
X_inner = X[y==0]
X_outer = X[y==1]
```

```python
plt.scatter(X_inner[:, 0], X_inner[:, 1], c="r")
plt.scatter(X_outer[:, 0], X_outer[:, 1], c="b")
plt.show()
```

```python
z = np.zeros(n)
#z = x^2+y^2
z_inner = -X_inner[:, 0] **2 - X_inner[:, 1]**2
z_outer = -X_outer[:, 0] **2 - X_outer[:, 1]**2
```

```python
fig = plt.figure(figsize = (8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_inner[:, 0], X_inner[:,1], z_inner, c='r')
ax.scatter(X_outer[:, 0], X_outer[:,1], z_outer, c='b')
```
