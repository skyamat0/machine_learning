```python
from sklearn.datasets import load_boston
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
boston = load_boston()
X = boston.data
Y = boston.target.reshape(-1, 1)
sscaler = StandardScaler()
sscaler_X = sscaler.fit_transform(X)
sscaler_Y = sscaler.fit_transform(Y)
```

```python
sscaler_X_train, sscaler_X_test, sscaler_Y_train, sscaler_Y_test = train_test_split(sscaler_X, sscaler_Y, random_state = 0)
```

```python
sscaler_svr = SVR(kernel='rbf', C=10, gamma='scale', epsilon=0.4)
sscaler_svr.fit(sscaler_X_train, sscaler_Y_train)
```

kernel{‘linear’, ‘poly’, ‘rbf’, ‘sigmoid’, ‘precomputed’}, default=’rbf’

```python
print(f"sscaler train accuracy: {sscaler_svr.score(sscaler_X_train, sscaler_Y_train)}")
print(f"sscaler test accuracy: {sscaler_svr.score(sscaler_X_test, sscaler_Y_test)}")
```
