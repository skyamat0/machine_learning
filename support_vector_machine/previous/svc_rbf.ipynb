```python
from sklearn.svm import SVC
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mglearn
from mpl_toolkits.mplot3d import Axes3D, axes3d
!pip install mglearn
```

```python
#xorについての分類
#xorとは x_1=1の時に、x_2を反転させる排他的論理和のこと
xor_input = np.array([[1, 1],
                [1, 0],
                [0, 1],
                [0, 0]])
xor_output = np.array([0, 1 , 1, 0])
```

```python
for i in range(4):
    if xor_output[i] == 0:
        plt.scatter(xor_input[i][0], xor_input[i][1], c="red")
    else:
        plt.scatter(xor_input[i][0], xor_input[i][1], c="blue")
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("XOR")
```

```python
#x3 = 1/(1 + np.exp(-(xor_input[:, 0] + xor_input[:, 1])))
def sigmoid(x, y):
    return 1/(1+np.exp(-(x+y)))
def relu(x, y):
    return np.max(1, x-y)
def RBF(x, y):
    return np.exp(-np.sqrt(2) * np.abs(x - y)**2)
x1 = xor_input[:, 0]
x2 = xor_input[:, 1]
r2 = np.sqrt(2)
x3 = RBF(x1, x2)

fig = plt.figure(figsize = (8, 8))
ax = fig.add_subplot(111, projection='3d')
for i, j in enumerate(xor_output):
    if j == 0:
        ax.scatter(x1[i], x2[i], x3[i], c='r')
    else:
        ax.scatter(x1[i], x2[i], x3[i], c='b')
```

```python
X = np.array([x1, x2, x3]).T
y = xor_output
svc = SVC(kernel="linear").fit(X,y)
svc.score(X,y)
```
