# SVM分類器を作成する

## 注意！！！
svmは基本的に標準化が必要．

アルゴリズムで「サポートベクター間の距離」が使われるため，標準化が必要になっている．



```python
from sklearn.svm import SVC
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
import numpy as np
km = np.array([10, 5])
g = np.array([60000, 80000])
np.linalg.norm(km-g)
```

上記の例だと，ほぼほぼgの情報しか入っていない．
kmに関する情報はこの距離には反映されていない．

```python
x = np.array([km, g]).T
sscaler = StandardScaler()
sscaler.fit(x)
x = sscaler.transform(x)
print(x)
np.linalg.norm(x)
```

これで，kmの情報とgの情報が距離に均等に反映されるようになった．

```python
cancer = load_breast_cancer()
X = cancer.data
y = cancer.target
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
# 訓練データを標準化しよう

# 標準化のためのインスタンスを生成
sscaler = StandardScaler() # WRITE ME

# データから平均や標準偏差を学習
# WRITE ME
sscaler.fit(X_train)

# 学習したsscalerを使ってデータを変換
X_train_scaled = sscaler.transform(X_train)# WRITE ME

# 同様のsscalerでテストデータを変換
X_test_scaled = sscaler.transform(X_test) # WRITE ME
```

```python
# 訓練データを使って学習

# svmのインスタンスを生成
args = {"C":1, "kernel":"linear"}
svc = SVC(**args) # WRITE ME

# svmモデルを学習
# WRITE ME
svc.fit(X_train_scaled, y_train)

# 精度を見る
print("訓練データ精度:", svc.score(X_train_scaled, y_train))
print("テストデータ精度:", svc.score(X_test_scaled, y_test))
```
