# サポートベクトルマシン（サポートベクタマシン）

サポートベクトルを用いて，線形分離を行うモデル．

サポートベクトル：分離境界に一番近いベクトルのこと

```python
from sklearn.svm import SVC
```

````python
# ライブラリのインポート
from sklearn.svm import SVC
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

````python
# データを定義
X_and = np.array([[1, 1],
                 [1, 0],
                 [0, 1],
                 [0, 0]])
y_and = np.array([1, 0, 0, 0])
plt.scatter(X_and[y_and==1][:, 0], X_and[y_and==1][:, 1], c="b")
plt.scatter(X_and[y_and==0][:, 0], X_and[y_and==0][:, 1], c="r")
plt.show()
````

````python
# インスタンスを生成し，学習させる
svc = SVC(C=10, kernel="linear")
svc.fit(X_and, y_and)
````

````python
# 比較のためにロジスティック回帰も実装する
from sklearn.linear_model import LogisticRegression

# 分離境界を描画して見る

# 描画用の配列を用意
x = np.linspace(0, 1, 101)
y = np.linspace(0, 1, 101)
xx, yy = np.meshgrid(x, y)
X_pred = np.hstack([xx.reshape(-1, 1), yy.reshape(-1, 1)])

# 用意した配列を用いて各点における予測値を得る．
y_pred = svc.predict(X_pred)
print("SVM score: ",svc.score(X_and, y_and))

# 各店とその予測値を描画
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].scatter(X_pred[y_pred==1][:, 0], X_pred[y_pred==1][:, 1], c="b")
ax[0].scatter(X_pred[y_pred==0][:, 0], X_pred[y_pred==0][:, 1], c="r")
ax[0].scatter(X_and[y_and==1][:, 0], X_and[y_and==1][:, 1], c="b")
ax[0].scatter(X_and[y_and==0][:, 0], X_and[y_and==0][:, 1], c="r")
ax[0].set_title("SVM")

# Logistic回帰と比較
lr = LogisticRegression(C=10,)
lr.fit(X_and, y_and)
print("Logistic Regression Score:", lr.score(X_and, y_and))
y_pred_lr = lr.predict(X_pred)
ax[1].scatter(X_pred[y_pred_lr==1][:, 0], X_pred[y_pred_lr==1][:, 1], c="b")
ax[1].scatter(X_pred[y_pred_lr==0][:, 0], X_pred[y_pred_lr==0][:, 1], c="r")
ax[1].set_title("Logistic Regression")
fig.show()
````
