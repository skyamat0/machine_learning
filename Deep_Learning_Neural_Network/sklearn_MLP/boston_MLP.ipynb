```python
from sklearn.datasets import load_boston
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
boston = load_boston()
X, Y = boston.data, boston.target
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=0)
```

```python
mlp = MLPRegressor(hidden_layer_sizes=(100, ),alpha=1, max_iter=10000, random_state=0)
mlp.fit(X_train, Y_train)
```

```python
print(f"train score : {mlp.score(X_train, Y_train)}")
print(f"test score : {mlp.score(X_test, Y_test)}")
```

```python
sscaler = StandardScaler()
sscaler.fit(X_train)
scaled_X_train = sscaler.transform(X_train)
scaled_X_test = sscaler.transform(X_test)
```

```python
mlp.fit(scaled_X_train, Y_train)
print(f"scaled train score : {mlp.score(scaled_X_train, Y_train)}")
print(f"scaled test score : {mlp.score(scaled_X_test, Y_test)}")
```
