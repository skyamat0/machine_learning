```python
import numpy as np

import matplotlib.pyplot as plt
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def tanh(x):
    return (np.exp(2 * x) - 1) / (np.exp(2 * x) + 1)
def relu(x):
    return np.maximum(x, 0)
def step(x):
    return np.array(x > 0, dtype=int)
```

```python
display(mglearn.plots.plot_logistic_regression_graph())
#logistic回帰
#y = w[0]x[0] + w[1]x[1] + w[2]x[2] + w[3]x[3]
#このyをシグモイド関数に押し込めて、0~1の間の範囲を出力した
display(mglearn.plots.plot_single_hidden_layer_graph())
```

```python
#よく使われる活性化関数
fig, axes = plt.subplots(1, 4, figsize=(15, 5))
axes[0].plot(np.arange(-5, 5, 0.1), tanh(np.arange(-5, 5, 0.1)))
axes[1].plot(np.arange(-5, 5, 0.1), sigmoid(np.arange(-5, 5, 0.1)))
axes[2].plot(np.arange(-5, 5, 0.1), relu(np.arange(-5, 5, 0.1)))
axes[3].plot(np.arange(-5, 5, 0.1), step(np.arange(-5, 5, 0.1)))
```
