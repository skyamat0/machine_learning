```python
from sklearn.datasets import make_moons
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
import mglearn
import matplotlib.pyplot as plt
```

```python
X, Y = make_moons(n_samples=100, noise=0.3, random_state=0)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=0)
```

```python
mlp_relu = MLPClassifier(max_iter=1000,
                         hidden_layer_sizes=(10, 10),
                         activation="relu",
                         solver="lbfgs",
                         random_state=0)
mlp_relu.fit(X_train, Y_train)
```

```python
mglearn.plots.plot_2d_separator(mlp_relu, X_train, fill=True, alpha=.3)
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], Y_train)
```

```python
mlp_relu_100 = MLPClassifier(max_iter=1000,
                         hidden_layer_sizes=(100, 100), 
                         activation="relu", 
                         solver="lbfgs",
                         random_state=0)
mlp_relu_100.fit(X_train, Y_train)
```

```python
mglearn.plots.plot_2d_separator(mlp_relu_100, X_train, fill=True, alpha=.3)
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], Y_train)
```

```python
mlp_sigmoid = MLPClassifier(max_iter=1000, 
                            hidden_layer_sizes=(10, 10), 
                            activation="logistic", 
                            solver="lbfgs",
                            random_state=0)
mlp_sigmoid.fit(X_train, Y_train)
```

```python
mglearn.plots.plot_2d_separator(mlp_sigmoid, X_train, fill=True, alpha=.3)
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], Y_train)
```

```python
mlp_tanh = MLPClassifier(max_iter=1000,
                         hidden_layer_sizes=(10, 10),
                         activation="tanh",
                         solver="lbfgs",
                         random_state=0)
mlp_tanh.fit(X_train, Y_train)
```

```python
mglearn.plots.plot_2d_separator(mlp_tanh, X_train, fill=True, alpha=.3)
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], Y_train)
```

```python
mlp_relu_10_5 = MLPClassifier(hidden_layer_sizes=(100, 100, 100, 100, 100),
                         activation="relu",
                         solver="lbfgs",
                         random_state=0)
mlp_relu_10_5.fit(X_train, Y_train)
```

```python
mglearn.plots.plot_2d_separator(mlp_relu_10_5, X_train, fill=True, alpha=.3)
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], Y_train)
```

```python
print(f"mlp_relu train score : {mlp_relu.score(X_train, Y_train)}")
print(f"mlp_relu test score : {mlp_relu.score(X_test, Y_test)}")
print(f"mlp_sigmoid train score : {mlp_sigmoid.score(X_train, Y_train)}")
print(f"mlp_sigmoid test score : {mlp_sigmoid.score(X_test, Y_test)}")
print(f"mlp_tanh train score : {mlp_tanh.score(X_train, Y_train)}")
print(f"mlp_tanh test score : {mlp_tanh.score(X_test, Y_test)}")
print(f"mlp_relu_100 train score : {mlp_relu_100.score(X_train, Y_train)}")
print(f"mlp_relu_100 test score : {mlp_relu_100.score(X_test, Y_test)}")
print(f"mlp_relu_10_5 train score : {mlp_relu_10_5.score(X_train, Y_train)}")
print(f"mlp_relu_10_5 test score : {mlp_relu_10_5.score(X_test, Y_test)}")
```
